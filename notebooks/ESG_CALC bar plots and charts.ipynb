```python

df_sed = pd.DataFrame(stage_emissions_data, columns=["stage_number", "diesel_consumption","co2_consumption", "ghg_consumption" ])
df_sed
```

```python
# reverse order, so stage 1 comes first

df_sed = df_sed.sort_values(by=['stage_number'])

# add dollar impact

carbon_credit = 30
diesel_factor = 0.01
ghg_factor = 1.24

df_sed['USD_carbon_cost'] = carbon_credit* (df_sed['co2_consumption'] + ghg_factor*df_sed['ghg_consumption'] + diesel_factor* df_sed['diesel_consumption'])

# add cumsum vals to df

df_sed['cumulative_co2_consumption'] = df_sed['co2_consumption'].cumsum()
df_sed['cumulative_diesel_consumption'] = df_sed['diesel_consumption'].cumsum()
df_sed['cumulative_ghg_consumption'] = df_sed['ghg_consumption'].cumsum()
df_sed['cumulative_USD_carbon_cost'] = df_sed['USD_carbon_cost'].cumsum()


df_sed
```

```python

figsize = (20,10)

# look at CO2 per stage

ax = df_sed.plot.bar(x='stage_number', y='co2_consumption',figsize=figsize, color='blue')

ax2 = ax.twinx()

ax2 = df_sed.plot(kind='line', x='stage_number', y='cumulative_co2_consumption', ax=ax2, color='red')

ax2.tick_params(axis='y', colors='r')
ax2.set_ylabel('cum co2, MTCO2', color = 'r')

ax.tick_params(axis='y', colors='b')
ax.set_ylabel('co2, MTCO2', color = 'b')

plt.show()
```
